```python
# ==== load & process three runs ====
ids = [57, 59, 60, 61, 62, 63, 64, 65, 66, 67]

# sampling parameters (unchanged)
T = 200e-6   # Sampling period (s)
Fs = 1/T     # Sampling frequency (Hz)
L = 800000   # Length of signal
t = np.linspace(0, L*T, L, endpoint=False)  # time vector

# frequency axis (unchanged)
frequencies = np.fft.fftfreq(L, T)
mask = frequencies >= 0

# compute FFT magnitudes for each id and each channel
all_fft_magnitudes = {}  # {id: {k: spectrum}}
for run_id in ids:
    signal = load_by_run_spec(captured_run_id=run_id).to_xarray_dataset()

    # load & mean-center (unchanged)
    signal_R, signal_Phi, signal_I, signal_Q = [
        signal[k] - np.mean(signal[k]) for k in ['R', 'Phi', 'I', 'Q']
    ]
    signal = {k: v - np.mean(v) for k, v in signal.items()}

    # FFT (unchanged)
    fft_magnitudes = {
        k: 2.0/L * np.abs(np.fft.fft(sig)[mask])
        for k, sig in zip(['R', 'Phi', 'I', 'Q'], [signal_R, signal_Phi, signal_I, signal_Q])
    }
    all_fft_magnitudes[run_id] = fft_magnitudes

# ==== plot: overlay all three runs on the same axes per channel ====
labels = {'R': 'Magnitude', 'Phi': 'Phase', 'I': 'I', 'Q': 'Q'}

for k, ylabel in labels.items():
    plt.figure(figsize=(10,6))
    for run_id in ids:
        plt.plot(frequencies[mask], all_fft_magnitudes[run_id][k], label=f'id {run_id}')
    plt.title(f'FFT of {k}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(ylabel)
    plt.xlim(0, Fs/2)   # show only 0..Nyquist for readability
    plt.grid()
    plt.legend()
    plt.show()
```
