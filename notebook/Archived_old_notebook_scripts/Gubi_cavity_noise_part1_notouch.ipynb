```python
# loading data
id = 57
signal = load_by_run_spec(captured_run_id=id).to_xarray_dataset()
```

```python
# sampling parameters
T = 200e-6   # Sampling period (s)
Fs = 1/T  # Sampling frequency (Hz)
L = 800000   # Length of signal
t = np.linspace(0, L*T, L, endpoint=False)  # time vector
```

```python
# loading signal
signal_R, signal_Phi, signal_I, signal_Q = [
    signal[k] - np.mean(signal[k]) for k in ['R', 'Phi', 'I', 'Q']
]
signal = {k: v - np.mean(v) for k, v in signal.items()}
```

```python
# Compute FFT
frequencies = np.fft.fftfreq(L, T)
mask = frequencies >= 0

fft_magnitudes = {
    k: 2.0/L * np.abs(np.fft.fft(sig)[mask])
    for k, sig in zip(['R', 'Phi', 'I', 'Q'], [signal_R, signal_Phi, signal_I, signal_Q])
}
```

```python
# 4. Plot the Results
labels = {'R': 'Magnitude', 'Phi': 'Phase', 'I': 'I', 'Q': 'Q'}

for k, ylabel in labels.items():
    plt.figure(figsize=(10,6))
    plt.plot(frequencies[mask], fft_magnitudes[k])
    plt.title(f'FFT of {k}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(ylabel)
    plt.grid()
    plt.show()
```
