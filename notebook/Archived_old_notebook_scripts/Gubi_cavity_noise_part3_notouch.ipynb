```python
# ==== map run IDs to frequencies ====
id_to_freq = {
    57: 4.3725e9,
    59: 4.3740e9,
    60: 4.3780e9,
    61: 4.3820e9,
    62: 4.3940e9,
    63: 4.4060e9,
    64: 4.3700e9,
    65: 4.3600e9,
    66: 4.3530e9,
    67: 4.3450e9,
}

# select run IDs
ids = list(id_to_freq.keys())

# reference frequency
ref_freq = 4.3725e9

# relative frequencies (Hz) w.r.t reference
relative_freqs = {rid: id_to_freq[rid] - ref_freq for rid in ids}

# ==== sampling parameters ====
T = 200e-6   # Sampling period (s)
Fs = 1/T     # Sampling frequency (Hz)
L = 800000   # Length of signal
t = np.linspace(0, L*T, L, endpoint=False)  # time vector

# frequency axis
frequencies = np.fft.fftfreq(L, T)
mask = frequencies >= 0
freq_axis = frequencies[mask]

# ==== compute FFT magnitudes for each run ====
all_fft_magnitudes = {}
for run_id in ids:
    signal = load_by_run_spec(captured_run_id=run_id).to_xarray_dataset()

    # load & mean-center
    signal_R, signal_Phi, signal_I, signal_Q = [
        signal[k] - np.mean(signal[k]) for k in ['R', 'Phi', 'I', 'Q']
    ]

    # FFT
    fft_magnitudes = {
        k: 2.0/L * np.abs(np.fft.fft(sig)[mask])
        for k, sig in zip(['R', 'Phi', 'I', 'Q'], [signal_R, signal_Phi, signal_I, signal_Q])
    }
    all_fft_magnitudes[run_id] = fft_magnitudes

# ==== plot: overlay all runs, ordered by relative frequency ====
labels = {'R': 'Magnitude', 'Phi': 'Phi', 'I': 'I', 'Q': 'Q'}

for k, ylabel in labels.items():
    plt.figure(figsize=(10,6))
    
    # sort ids by relative frequency (negative  positive)
    sorted_ids = sorted(ids, key=lambda rid: relative_freqs[rid])
    
    for run_id in sorted_ids:
        rel_freq = relative_freqs[run_id]  # offset from reference
        plt.plot(freq_axis,
                 all_fft_magnitudes[run_id][k],
                 label=f'{rel_freq/1e6:+.1f} MHz')  # relative in MHz
    
    plt.title(f'FFT of {k}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(ylabel)
    plt.xlim(0, Fs/2)   # show only 0..Nyquist
    plt.grid()
    plt.legend()
    plt.show()
```
